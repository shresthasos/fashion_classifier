```python
import tensorflow as tf
```

```python
from tensorflow import keras
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
keras.utils.set_random_seed(42)
```

```python
(X_train, y_train) ,(x_test, y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
labels = [
    "T-shirt/top",
    "Trouser",
    "Pullover",
    "Dress",
    "Coat",
    "Sandal",
    "Shirt",
    "Sneaker",
    "Bag",
    "Ankle boot"
]
```

```python
x_train = X_train/255.0
x_test = x_test /255.0
```

```python
x_train.shape
```

```python
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test,-1) #expanding for chhanels dimensions
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
input_Shape = x_train.shape[1:]
```

```python
input_Shape
```

```python
from numpy._core.fromnumeric import shape
#convolutional layer
input = keras.Input(shape=input_Shape)
x = keras.layers.Conv2D(32, kernel_size=(2,2), activation="relu", name="Conv_1")(input)

#pooling layer
x = keras.layers.MaxPool2D()(x)
```

```python
x = keras.layers.Conv2D(32, kernel_size=(2,2), activation="relu", name="Conv_2")(x)
```

```python
x = keras.layers.MaxPool2D()(x)
```

```python
x = keras.layers.Flatten()(x)
```

```python
x = keras.layers.Dense(256, activation ="relu")(x)
```

```python
output = keras.layers.Dense(10, activation ="softmax")(x)
model = keras.Model(input, output)
```

```python
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
history = model.fit(x_train , y_train, batch_size=64, epochs = 10, validation_split=0.2)
```

```python
def plot_loss_curves(history):
  plt.clf()
  history_dict = history.history
  loss_values = history_dict["loss"]
  val_loss_values = history_dict["val_loss"]
  epochs = range(1, len(loss_values) + 1)
  plt.plot(epochs, loss_values, "bo", label="Training loss")
  plt.plot(epochs, val_loss_values, "b", label = "Validation loss")
  plt.title("Training and validation loss")
  plt.xlabel("Epochs")
  plt.ylabel("loss")
  plt.show()
```

```python
plot_loss_curves(history)
```

```python
def plot_acc_curves(history):
  plt.clf()
  history_dict = history.history
  acc = history_dict["accuracy"]
  val_acc = history_dict["val_accuracy"]
  epochs = range(1, len(acc) +1)
  plt.plot(epochs, acc, "bo", label="Training acc")
  plt.plot(epochs, val_acc, "b", label="validation acc")
  plt.title("Training and validation accuracy")
  plt.xlabel("epochs")
  plt.ylabel("Accuracy")
  plt.legend()
  plt.show()
```

```python
plot_acc_curves(history)
```

```python
model.evaluate(x_test,y_test)
```
